# `pandas` library exploration

```python
# Import module
import pandas as pd
```

```python
# Create dataframe from CSV (from lab 5 exercise)
drivers_df = pd.read_csv('lab05-files/drivers.csv')
results_df = pd.read_csv('lab05-files/results.csv')

print(drivers_df)
print(results_df)
```

```python
# Drop certain columns from dataframe
drivers_df.drop(['driverRef', 'number', 'dob', 'nationality', 'url'], inplace=True, axis=1)
results_df.drop(['raceId', 'constructorId', 'number', 'grid', 'position', 'positionText', 'points', 'laps', 'time', 'milliseconds', 'fastestLap', 'rank', 'fastestLapTime', 'fastestLapSpeed', 'statusId'], inplace=True, axis=1)

print(drivers_df)
print(results_df)
```

```python
# Filter rows for first place
results_df = results_df[results_df['positionOrder'] == 1]

print(results_df)
```

```python
# Merge dataframes
merged_df = results_df.merge(drivers_df, on='driverId', how='left')

print(merged_df)
```

```python
# Sort dataframe
merged_df = merged_df.sort_values(by=["driverId"])

print(merged_df)
```
