# `sys` library exploration

```python
# Import module
import sys
```

```python
# Get command line arguments
args = sys.argv
print(args)
print(args[0])
```

```python
# Get command line flags
flags = sys.flags
print(flags.verbose)
```

```python
# Get current platform
print(sys.platform)
```

```python
# Get interpreter version
print(sys.version)
```
